### descripcion del problema
Sin POO, tendrías que manejar los datos de cada banco por separado (variables sueltas), lo cual genera desorden y repetición de código.

La clase Banco resuelve esto mediante:

Encapsulamiento: Agrupa los datos (atributos) y acciones (funciones) en un solo "molde".

Escalabilidad: Te permite crear cientos de bancos distintos (banco1, banco2, etc.) usando el mismo código, manteniendo cada uno su información independiente y organizada.

```python
class Banco:
    def __init__(self, nombre, numeroSucursales, activosTotales, calificacionRiesgo):
        self.nombre = nombre
        self.numeroSucursales = numeroSucursales
        self.activosTotales = activosTotales
        self.calificacionRiesgo = calificacionRiesgo

    def otorgarCredito(self, monto):
        return f"-> [ACCIÓN]: {self.nombre} aprobó un crédito por ${monto}."

    def invertirFondos(self, mercado):
        return f"-> [ACCIÓN]: {self.nombre} movió capital al mercado de {mercado}."

    def emitirTarjeta(self, tipo):
        return f"-> [ACCIÓN]: {self.nombre} entregó una tarjeta {tipo} al cliente."

# 1. Creamos los objetos
banco1 = Banco("Banco Azteca", 1200, 500000000, "A+")
banco2 = Banco("BBVA", 800, 950000000, "AA")

# 2. Interacción con el usuario
print("--- SISTEMA BANCARIO POO ---")
seleccion = input(f"¿Qué banco quieres gestionar? (1: {banco1.nombre} / 2: {banco2.nombre}): ")

banco_actual = banco1 if seleccion == "1" else banco2

print(f"\nHas seleccionado: {banco_actual.nombre}")
print(f"Sucursales: {banco_actual.numeroSucursales} | Riesgo: {banco_actual.calificacionRiesgo}")

while True:
    print("\n¿Qué deseas hacer?")
    print("1. Otorgar Crédito")
    print("2. Invertir Fondos")
    print("3. Emitir Tarjeta")
    print("4. Salir")

    opcion = input("Elige una opción (1-4): ")

    if opcion == "1":
        monto = input("Ingresa el monto del crédito: ")
        print(banco_actual.otorgarCredito(monto))
    elif opcion == "2":
        mercado = input("¿En qué mercado invertir (ej. Acciones, Oro, Cripto)?: ")
        print(banco_actual.invertirFondos(mercado))
    elif opcion == "3":
        tipo = input("¿Qué tipo de tarjeta (ej. Débito, Crédito, Platino)?: ")
        print(banco_actual.emitirTarjeta(tipo))
    elif opcion == "4":
        print("Saliendo del sistema...")
        break
    else:
        print("Opción no válida.")

print("programa terminado por jose solis NL 51")
```
